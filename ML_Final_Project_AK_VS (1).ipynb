```python
import numpy as np

# Constants for max values
MAX_DAY = 31
MAX_MONTH = 12
MAX_WDAY = 7
MAX_WEEK_OF_YR = 52

# Sine and Cosine transformations for 'intermediate_dataset'
intermediate_dataset['sin_day'] = np.sin(2 * np.pi * intermediate_dataset['day'] / MAX_DAY)
intermediate_dataset['cos_day'] = np.cos(2 * np.pi * intermediate_dataset['day'] / MAX_DAY)

intermediate_dataset['sin_month'] = np.sin(2 * np.pi * intermediate_dataset['month'] / MAX_MONTH)
intermediate_dataset['cos_month'] = np.cos(2 * np.pi * intermediate_dataset['month'] / MAX_MONTH)

intermediate_dataset['sin_wday'] = np.sin(2 * np.pi * intermediate_dataset['wday'] / MAX_WDAY)
intermediate_dataset['cos_wday'] = np.cos(2 * np.pi * intermediate_dataset['wday'] / MAX_WDAY)

intermediate_dataset['sin_week_of_yr'] = np.sin(2 * np.pi * intermediate_dataset['week_of_yr'] / MAX_WEEK_OF_YR)
intermediate_dataset['cos_week_of_yr'] = np.cos(2 * np.pi * intermediate_dataset['week_of_yr'] / MAX_WEEK_OF_YR)

# Displaying the updated DataFrame
print(intermediate_dataset)
```

```python
columns_to_front = ['day', 'month', 'year', 'wday', 'week_of_yr', 'event_type_1','event_name_1','sin_day','cos_day','sin_month','cos_month','sin_wday','cos_wday','sin_week_of_yr','cos_week_of_yr']

remaining_columns = [col for col in intermediate_dataset.columns if col not in columns_to_front]

# Concatenate the lists to form the new column order
new_column_order = columns_to_front + remaining_columns

# Reindex the DataFrame with the new column order
intermediate_dataset = intermediate_dataset[new_column_order]

print(intermediate_dataset.columns)  # Just to verify the new order
```

```python
final_dataset = intermediate_dataset.drop(columns = ["day","month","year","wday","week_of_yr"])
final_dataset.head()
```

```python
generic_params = final_dataset.iloc[:,:10]
print(generic_params)
other_params = final_dataset.iloc[:,10:]
print(other_params)
```

```python
item_store_list = other_params.columns.tolist()
factor_list = generic_params.columns.tolist()
print(factor_list)
print(len(item_store_list))
```

```python
def df_to_X_y(df, window_size = 25):
    df_as_np = df.to_numpy()
    X = []
    y = []
    for i in range(len(df_as_np)-window_size):
        row = [r for r in df_as_np[i:i + window_size]]
        X.append(row)
        label = df_as_np[i+window_size][0]
        y.append(label)
    X = np.array(X).astype('float32')
    y = np.array(y).astype('float32')
    return X, y
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import Sequential
from tensorflow.keras import layers
from keras.layers import LSTM
from tensorflow.keras import optimizers
from tensorflow.keras.callbacks import ModelCheckpoint
from tensorflow.keras.optimizers import Adam, RMSprop
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.metrics import RootMeanSquaredError
```

```python
model = Sequential()
model.add(layers.InputLayer((25,11)))
model.add(LSTM(64, return_sequences = True))
#model.add(layers.Dropout(0.2))
model.add(LSTM(32,return_sequences = False))
model.add(layers.Dense(16,'relu'))
model.add(layers.Dense(1, 'linear'))

model.summary()
```

```python
cp = ModelCheckpoint('model/', save_best_only = True)
model.compile(loss = MeanSquaredError(), optimizer = Adam(learning_rate = 0.0001), metrics = [RootMeanSquaredError()])
```

```python
import matplotlib.pyplot as plt

def plot_predictions(model, X, y, start=0, end=200):
  predictions = model.predict(X)
  q_preds = predictions[:, 0]
  q_actuals = y
  df = pd.DataFrame(data={'Quantity Predictions': q_preds,
                          'Quantity Actuals':q_actuals
                          })
  plt.plot(df['Quantity Predictions'][start:end])
  plt.plot(df['Quantity Actuals'][start:end])
  return df[start:end]
```

```python
all_predictions = []

all_actuals = []

for i in range(len(item_store_list)):
    column_to_front = [item_store_list[i]]
    remaining_columns = factor_list
    new_column_order = column_to_front + remaining_columns
    final_data_subset = final_dataset[new_column_order]
    X, y = df_to_X_y(final_data_subset)
    X_train, y_train = X[:1530],y[:1530]
    X_val, y_val = X[1530:1800], y[1530:1800]
    X_test, y_test = X[1800:], y[1800:]
    quantity_train_mean = np.mean(X_train[:,:,0])
    quantity_train_std = np.std(X_train[:,:,0]) # make change here

    def preprocess(X):
        if quantity_train_std != 0:
            X[:,:,0] = (X[:,:,0] - quantity_train_mean)/quantity_train_std
        else:
            X[:,:,0] = 0
        
    
    preprocess(X_train)
    preprocess(X_val)
    preprocess(X_test)
    
    model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs = 30, callbacks = [cp])
    
    y_pred = model.predict(X_test)
    
    predictions = y_pred.tolist()
    
    all_predictions.append((item_store_list[i], predictions))
    
    all_actuals.append((item_store_list[i],y_test.tolist()))
    
    print(f"Round {i+1} Complete")
```

```python
rows_actual = []

columns=["Unique_ID"] + [f"Day_{i+1}" for i in range(28)]

for row in all_actuals:

    unique_id = row[0]
    last_28_values = row[1][-28:]

    item_row = [unique_id] + [r for r in last_28_values]
    
    rows_actual.append(item_row)

output_actuals = pd.DataFrame(rows_actual, columns = columns)

print(output_actuals.head())
```

```python
columns = ['Unique_ID'] + [f'F{i}' for i in range(1, 29)]
rows = []

for item in all_predictions:
    unique_id, values = item
    # Flatten the list of lists and take the last 28 values
    flattened_values = [round(value[0]) for value in values][-28:]
    row = [unique_id] + flattened_values
    rows.append(row)

# Create a DataFrame
output_preds_rounded = pd.DataFrame(rows, columns=columns)
```

```python
historical_diffs = historical_dataset.filter(like='d_').diff(axis=1).dropna(axis=1) ** 2

historical_diffs = historical_diffs.reset_index()

historical_diffs = historical_diffs.drop(columns = ['index'], axis = 1)

print(historical_diffs)
```

```python
output_actuals_aligned = output_actuals.filter(like='Day_')
output_preds_aligned = output_preds_rounded.filter(like='F')

# Optionally, rename columns to match
output_preds_aligned.columns = output_actuals_aligned.columns

output_actuals_filled = output_actuals_aligned.fillna(0)
output_preds_filled = output_preds_aligned.fillna(0)

squared_errors = (output_actuals_filled - output_preds_filled) ** 2

print(squared_errors)
```

```python
sigma = np.sqrt(historical_diffs.mean(axis=1)) + 1e-6

total_squared_errors = squared_errors.sum(axis=1)

normalized_squared_errors = total_squared_errors / sigma

n = squared_errors.shape[1]  # Number of forecasted days

rmsse = np.sqrt(normalized_squared_errors / n)
```

```python
total_squared_errors = squared_errors.sum().sum()  # Summing across all items and days
total_historical_diffs = historical_diffs.sum().sum()  # Summing across all items and days

# Calculate the number of observations used for prediction
n = squared_errors.shape[1]  # Number of days for which predictions were made

# Compute the mean of the historical differences
mean_historical_diffs = total_historical_diffs / (historical_diffs.shape[1] * (historical_diffs.shape[0] - 1))

# Calculate RMSSE
overall_rmsse = np.sqrt(total_squared_errors / (n * mean_historical_diffs))

print("Overall RMSSE:", overall_rmsse)
```
